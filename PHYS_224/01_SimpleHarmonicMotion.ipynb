# Harmonic Motion

## Properties

- harmonic motion has to be sonusoidal
- there must be a force tied to the motion

The spring and mass model is a good example.

### Position

> __Remeber Hookes Law: $f = -Kx$__

The force of a spring is a restoring force, which means that it is always trying to restore it to equilibrium.
The force always points towards the equilibrium point, and the displacement points opposite force

Position is given with $x(t) = Acos(wt)$ where $w$ is in rads/s. But since the position repeats itself every period we can make position a function of time and period, such that $x(t) = x(t + T)$. Setting each equation equal to each other, we get

\begin{equation}
A \cos (wt) =A \cos (w (t + T))\\
\cos (wt) = \cos (w(t + T))\\
\end{equation}


### Velocity and Acceleration

Velocity oscillates bewteen certain intervals.
Max velocity at equilibrium and zero velocity at turning points. This is because of an exchange of energy between potential and kinetic energy.

We can derive velocity and acceleration from position in terms of $w$ and $t$

\begin{equation}
x(t) = A \cos (wt)\\
v(t) = \frac{dx}{dt} = -Aw \sin (wt)\\
a(t) = \frac{d^2x}{dt^2} = (Aw)^2 \cos (wt) = w^2 x(t)\\
\end{equation}

We can see that we can relate acceleration to position through $w$ in that

\begin{equation}
\frac{d^2x}{dt^2} = w^2 x\\
\textrm{where}\\
w^2 = \frac{k}{m}\\
\textrm{or}\\
x = \frac{2 \pi}{T}\\
\textrm{where } T = period = \frac{1}{f}\\
\end{equation}

This equation is the universal equation of a simple harmonic oscillator in an ideal system

## Partial and General Solutions

When we talk about solutions to th simple harmonic motion problem, we want to find the solution describing position which satisfies the ordinary, first order, linear, homogeneous differential equation relating acceleration and position.

\begin{equation}

F = ma = -kx\\
m \frac{d^2x}{dt^2} = -kx\\
m \frac{d^2x}{dt^2} + kx = 0\\
\end{equation}

Since this is equation is ordinary, there is guaranteed to be one unique solution. The old equation we used to describe position $x(t) = A \cos (wt)$ was only a partial solution because we couldn't describe all possible scenarios with them. We can use the equation $x(t) = A \cos (wt + \phi)$ to describe the general solution because it has the $\phi$ parameter which describes the phase angle that we can use to describe any scenario that we can encounter in the SHO.

The $A$ and $\phi$ paramters are given in the initial conditions of the SHO problem, whereas, the $w$ parameter is the natural frequency of oscillation and will be specific to the properties of the pring chosen. This has important implications to SHO. For example, the period of the swing of a pendulum is not affected by the amplitude that it can reach. It tells us that since $w = \sqrt{k/m}$, the angular frequency increases as the displacement increases, and decreases as the mass increases.





### Worked Example

Suppose we have a SHO problem with the following properties

\begin{equation}
m = 0.8 kg\\
k = 180 N m^-1\\
x_0 = 0.04 m\\
v_0 = -0.5 \frac{m}{s}\\
\end{equation}

Find the expression for the displacement in the form of $$x = A \cos (wt + \phi)$$

We can first use $w = \sqrt{\frac{k}{m}} = \sqrt{\frac{180}{0.8}} = 15$ to find the angular frequency of the problem.

Then we can plug this in into the position and velocity functions along with the initial conditions hopin to find the final equations that include $\phi$ and $A$

\begin{equation}
x = 0.04 = A \cos (w(0) + \phi)\\
0.04 = A \cos(\phi)\\
0.0016 =A^2 \cos(\phi) 
\end{equation}

\begin{equation}
v = -0.5 = A(15) \sin (w(0) + \phi)\\
0.5 = -15A \sin(\phi)\\
\frac{0.25}{225} = A^2 \sin(\phi)\\


\textrm{Adding these two equations together, we get}\\
0.0016 + \frac{0.25}{225} = A^2 (\cos^2(\phi) + \sin^2(\phi))\\

A = \sqrt(0.0016 + \frac{0.25}{225}) = 0.05206 m\\
\end{equation}

Now we can plug $A$ to find $\phi$

\begin{equation}
0.04 = 0.05206 \cos(\phi)\\
\phi = \arccos(\frac{0.04}{0.05206})\\
\phi = 	0.6945 \textrm{ radians or } 39.8 \textrm{ degrees}

\end{equation}

## Energy of a Simple Harmonic Oscillator

We know that total energy $E$ is made up of kinetic energy $K$ and potential energy $U$, so 

\begin{equation}
E = K + U\\
E = \frac{1}{2}mv^2 + \frac{1}{2}kx^2\\

\textrm{We can substitute the previously derived equations for position and velocity}\\

E = \frac{1}{2}m(-Aw \sin(wt + \phi))^2 + \frac{1}{2}k(A \cos(wt + \phi))^2\\

E = \frac{1}{2}mA^2w^2 \sin^2(wt + \phi) + \frac{1}{2}kA^2 \cos^2(wt + \phi)\\

\textrm{And replacing $w^2$ with $\frac{k}{m}$, we get}\\

E = \frac{1}{2}kA^2 \sin^2(wt + \phi) + \frac{1}{2}kA^2 \cos^2(wt + \phi)\\

E = \frac{1}{2}kA^2 [\sin^2(wt + \phi) + \cos^2(wt + \phi)]\\

E = \frac{1}{2}kA^2\\


\end{equation}


This shows us that the total energy of a simple harminoc oscillator is proportional to the square of the Amplitude of the oscillator. This also shows us that the energy is conserved between potential and kinetic energy

## The physics of small vibrations

_Most oscillating systems will oscillate with SHM when the amplitude of oscillation is small._

This is because the potential energy plot of a SHO is parabolic shaped, and when the amplitude of a non SHO is small the plot of the potential energy of this non SHO will match closely to tht of the SHO potential energy plot.

Taylor approximations state that any function thich is continuous and differentiable at all times can becexpanded in a power series (x-a), when x is approximately a. 

\begin{equation}

f(x) = f(a) + \frac{(x-a)}{1!} (\frac{df}{dx}) + \frac{(x-a)^2}{2!} (\frac{d^2f}{dx^2}) + \dots + \frac{(x-a)^n}{n!} (\frac{d^nf}{dx^n})\\

\textrm{Therefore when we want to evaluate a small vibration about a = 0, we can use}\\

U(x) = U(0) + x (\frac{df}{dx}) + \frac{x^2}{2} (\frac{d^2U}{dx^2}) + \dots + \frac{x^n}{n!} (\frac{d^nU}{dx^n})
\end{equation}

However, we can eliminate the first term because potential energy is relative to any point. And since the first term is constant we can set the point for zero potential energy at x = 0, and since the first term is a constant, the value won't change as x changes. And the second term can be eliminated because if the potential energy is at a minimum at $x = 0$, then $\frac{dU}{dx}$ will be 0.

So we can take a good approximation when x is near 0 by taking the first non zero term of this expansion

$$U(x) = \frac{x^2}{2} (\frac{d^2U}{dx^2}) \\$$

We can use this approximation to derive an approximation of the $F_{sp}$.

\begin{equation}
F_{sp} = - \frac{dU}{dx} = \frac{d}{dx} \frac{x^2}{2} (\frac{d^2U}{dx^2}) = -x (\frac{d^2U}{dx^2})\\

\textrm{It is important to note that $\frac{d^2U}{dx^2}$ is a constant and therefore does not require a derivative calculation.}\\

\end{equation}

## Pendulums

### Simple Pendulums

In simple pendulims we have a point mass $m$, suspended from a massless, frictionless rod of length $l$. Given a displacement $\theta$, the arc length completed through $\theta$ is $l\theta$, the angular velocity along the arc is $l \frac{d\theta}{dt}$, the angular acceleration is $l\frac{d^2\theta}{dt^2}$.

<div style = "text-align: center"><img src="Pictures/simple_pendulum.png"></div>


We can see that the tensile forece of the rod cancels out with the radial force of gravity, leaving the restorative force of gravity to only be comrised of the tangential component $-mg\sin (\theta)$ We can use newtons second law to say

\begin{equation}
F = ma\\

-mg \sin (\theta) = ml\frac{d^2\theta}{dt^2}\\

\frac{d^2\theta}{dt^2} = - \frac{g}{l} \sin (\theta)\\

\end{equation}

This is the equation for simple harmonic motion for pendulums, however it differs from that of spring and mass systems $\frac{d^2}{dx^2} = -w^2x$ in that we have a $\sin (\theta)$ in the right hand side. However, we know that for small angles, we can approximate $\sin (\theta)$ as $\theta$. So we can say that as an approximation, the folling is true

$$\frac{d^2\theta}{dt^2} = - \frac{g}{l} \theta\\$$

And we can draw parallels from this equation to those of the mass and spring simple harmonic oscillators. We can say that $w = \sqrt{\frac{g}{l}}$, and $T = 2 \pi \sqrt{\frac{l}{g}}$, which then allows us to write formulas for angular displacement $\theta(t)$ and angular velocity $\frac{d\theta}{dt}$.

\begin{equation}
\theta(t) = \theta_0 \cos (wt + \phi)\\
\frac{d \theta}{dt} = - \theta_0 w \sin (wt + \phi)\\

\textrm{Where $\theta_0$ is the amplitude of the pendulum}\\
\end{equation}

We can see that period is independend of amplitude, which is why pendulum clocks made for good time keepers.



### Energy of Simple Pendulums

Here we want to analyze the kinetic and potential energies of a pendulum along its arc. We can say thay the displacement in the $y$ direction is $y = l\sin(\theta)$, which is approximately 

<div style = "text-align: center"><img src="Pictures/simple_pendulum_disp.png"></div>

Using pythagoreans theorem, we can caluclate the geometry of the pendulums displacement as

\begin{equation}
l^2 = (l-y)^2 + x^ 2\\
\textrm{Which gives}\\
2ly = x^ 2 + y^2\\
\end{equation}

However, since we are assuming small displacements for $\theta$, we know $x << l$, which means $y << x$, which means that we can ignore the $y^2$ term, giving us
$$y = \frac{x^2}{2l}\\$$

We can caluclate the total energy of the system as 

$$E = K_E + U_E = \frac{1}{2}mv^2 + mgy \\= \frac{1}{2}mv^2 + mg \frac{x^2}{2l}\\$$


When the pendulum is at its turning point, then all of its energy is in the form of potential energy, and its displacement x is equal to the amplitude A, so 

\begin{equation}
E = P_E = mg\frac{A^2}{2l}\\

\textrm{So we can say}\\

mg\frac{A^2}{2l} = \frac{1}{2}mv^2 + mg \frac{x^2}{2l}\\

\textrm{Which we can derive the velocity equation from once again}\\

\\v = \sqrt{\frac{g(A^2 - x^2)}{l}}\\

\\\textrm{Which we can integrate and once again derive the displacement function of by substituing x for $A\sin(\theta)$}\\

x = Asin(\sqrt{\frac{g}{l}}t + \phi)
\end{equation}


The energy equation for a mass on a spring is $E = \frac{1}{2}mv^2 + \frac{1}{2}kx^2$

The energy equation for a simple pendulum is $E = \frac{1}{2}mv^2 + \frac{1}{2} \frac{mg}{l} x^2$ 

We can see thay they both have equations of the form $E = \frac{1}{2} \alpha v^2 + \frac{1}{2} \beta x^2$. This is the character quality of simple harmonic oscillators. We can expand upon this by saying that $\alpha$ represents the inertia of the oscillator and $\beta$ represents the resroting force per unit displacement through which the system can store potential energy.


### Physical Pendulums

Physical pendulums differ form simple pendulums in that in physical pendulums, the mass is dispersed over the entire body of the pendulum, rather than being treated as a point mass with simple pendulums.

<div style="text-align: center"><img src="Pictures/physical_pendulum.png"></div>

If we have a rod of length $l$ and mass $m$, then its equation for the second lay goes from 
\begin{equation}
F = ma \\
\textrm{to}\\
I \frac{d^2 \theta}{d \theta ^2} = \tau\\

\textrm{Where $I$ is the moment of inertia of the rod, and $\tau$ is the applied torque from the center of mass of the rod.}\\
\end{equation}

Remember that moment of inertia os a uniform rof of lenghth $l$ and mass $m$ is $I = \frac{1}{3}ml^2$, and its center of mass is located at its midpoint. We can calculate the torque by using $$\tau = F \cdot D$$ where F is the force being appplied perpendicular to the rod, and D is the distance from the fulcrum to the midpoint, so

$$\tau = (\frac{1}{2}l) \times (-mg\sin(\theta))$$

Plugging back into the new equation for the second law we get

\begin{equation}
\frac{1}{3}ml^2\frac{d^2\theta}{dt^2} = -\frac{1}{2}mgl\sin(\theta)\\
\frac{d^2\theta}{d^2t} = -\frac{3g}{2l}\sin(\theta)\\
\textrm{However, with small angle approximation for $\sin(\theta)$, we can say}\\

\frac{d^2\theta}{dt^2} = -\frac{3g}{2l}\theta
\end{equation}

This gives us a simple hamonic oscillator with $w = \sqrt{3g/2l}$ and $T = 2\pi \sqrt{2l/3g}$
