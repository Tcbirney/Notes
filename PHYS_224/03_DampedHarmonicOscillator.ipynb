# Damped Harmonic Oscillator

A damped harmonic oscillator loses energy steadily at time moves on. The energy of the oscilator is directly tied to the displacement of the oscillator, and if we are losing energy of an oscillator over time, then we are also losing amplitude of the oscilator since the angular frequency of the oscillator stays constant.

The equation for displacement for a damped harmonic oscillator is
$$x(t) = A_0e^{-\beta t}\cos (\omega t + \phi)$$
Where $A_0$ is the initial amplitude, $\beta$ is the damping factor, and \omega is the angular frequency with damping, which is slightly different from the natural angular frequency, $w_0$.  

## Equations for a Damping Oscillator

The damping force is prortional to the negative velocity. So $F_d = -bv$. The faster we go, the more the damping force opposes us. Plugging this in to the 2nd law equation gives us.

Assume that we are on a horizontal spring, going right, with damping.

\begin{equation}
F= ma\\
F_{sp} - F_d  = ma\\
0 = ma + F_d - F_sp\\
0 = m \frac{d^2x}{dx^2} + b \frac{dx}{dt} + kx\\
0 = \frac{d^2x}{dx^2} + \frac{b}{m} \frac{dx}{dt} + \frac{k}{m}x\\
\textrm{We introduce the parameters}\\

w_o^2 = \frac{k}{m}, \gamma = \frac{b}{m} \textrm{ which are both positive}\\
\textrm{Which gives}\\

0 = \frac{d^2x}{dx^2} + \gamma \frac{dx}{dt} + w_o^2x\\


\end{equation}

## Light Damping

If we take the psoition eqaution we derived earlier $x(t) = Ae^{i(wt + \phi)}$, we need to know that this still needs to apply when we have no damping. We need to find a relation between $w$ and $w_o$. We also have the implicit inference that we are only considering the real component of this equation.

Lets derive velocity and accelteration

\begin{equation}
v(t) = iwAe^{i(wt + \phi)}\\
a(t) = (iw)^2Ae^{i(wt + \phi)} = -w^2Ae^{i(wt + \phi)}\\
\end{equation}

Lets plug these back into our charcateristic differential equation, and see what happens.

\begin{equation}
0 = \frac{d^2x}{dx^2} + \gamma \frac{dx}{dt} + {w_o}^2x\\
0 = -w^2Ae^{i(wt + \phi)} + \gamma iwAe^{i(wt + \phi)} + w_o^2 Ae^{i(wt + \phi)}\\
\textrm{We can extract some common factors and simplify}\\
0 = Ae^{i(wt + \phi)} (-w^2 + i\gamma w + w_o^2)
\end{equation}

We can see that there is a quadratic relationship if we use the damed frequency $w$ such that
\begin{equation}
w = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}\\
w = \frac{-i\gamma \pm \sqrt{(i\gamma)^2 - 4(-1)(w_o^2)}}{-2}\\
w = \frac{-i\gamma \pm \sqrt{4w^2 - \gamma^2}}{-2}\\
w = i \frac{i\gamma}{2} \pm \sqrt{\frac{4w_o^2}{4} - \frac{\gamma^2}{4}}\\
\tilde{w} = i \frac{\gamma}{2} \pm \sqrt{w_o^2 - (\frac{\gamma}{2})^2}\\

\end{equation}

__Light Damping is the case where $w_o^2 - (\frac{\gamma}{2})^2 > 0$__

The solution to this equation is what makes the charcateristic eqaution fo a lightly damped harmonic oscillator true.

So our solution for displacement becomes 
\begin{equation}  

x(t) = Ae^{i(wt + \phi)}\\
x(t) = Ae^{i([i \frac{\gamma}{2} \pm \sqrt{w_o^2 - (\frac{\gamma}{2})^2}]t + \phi)}\\
x(t) = Ae^{-i(\frac{\gamma}{2})t)} e^{\pm i \sqrt{w_o^2 - (\frac{\gamma}{2})^2} + \phi}\\
\textrm{since we are only considering the real component we can ignore the $\pm$ since $\cos (\phi) = \cos(-\phi)$}\\
\end{equation}

We can consider the $Ae^{-i(\frac{\gamma}{2})t)}$ as a time dependent amplitude, which makes sense because we can see that it has exponential decay since the damping force is a non restorative force that drains energy from the oscillator as time moves forwards. We can also seet hat this equation is extensible to the case ehre there is no damping force present. When no damping force, $\gamma$ is 0, so the exonential component of the amplitude is a constant 1, and the damped harmonic frequency becomes the same as the natural frequency.

We must assume that $w_o^2 - (\frac{\gamma}{2})^2 > 0$ so that we don't get any imaginary components in the exponential. This takes the form of $x(t) = Re(i\theta)$, which can also be written as $x(t) = R\cos(\theta) + i sin(\theta)$, where $R = Ae^{-\frac{\gamma}{2}t}$ and $\theta = \sqrt{w_o^2 - (\frac{\gamma}{2})^2} + \phi$. But since we agreed to only take the real compinent, we can ignore the $i\sin(\theta)$. So we can say that


$$x(t) = Ae^{-\frac{\gamma}{2}t} \cos(\sqrt{w_o^2 - (\frac{\gamma}{2})^2}t + \phi)$$

## Heavy Damping

__Heavy damping is the case where $w_o^2 - (\frac{\gamma}{2})^2 < 0$__

This means we need to have complex numbers in the derivation for $w$

\begin{equation}
\textrm{We have the derivation}\\
w = i\frac{\gamma}{2} + \sqrt{w_o^2 - (\frac{\gamma}{2})^2}\\
w = i\frac{\gamma}{2} + \sqrt{-((\frac{\gamma}{2})^2 - w_o^2)}\\
\textrm{If we say that $\alpha^2 = (\frac{\gamma}{2})^2 - w_o^2$, then}\\
w = i\frac{\gamma}{2} + \sqrt{-\alpha^2} = i\frac{\gamma}{2} \pm i\alpha\\
\textrm{But since we're only condisdering the real components of the $w$, then we can say}\\
w = \pm i\alpha
\end{equation}

We need to keep both the plus and minus sign versions f the solution becatse of the principle of uniqueness. This means that the complete solution of this oscillator will be a linear combination of the general solution for each of these versions of $w$. So we have

\begin{equation}
x(t) = Ae^{-\frac{\gamma}{2}t}e^{i(i\alpha t + \phi)} + Be^{-\frac{\gamma}{2}t}e^{i(-i\alpha t+ \phi)}\\
x(t) = Ae^{-\frac{\gamma}{2}t} e^{i^2\alpha t} e^{i\phi} + Be^{-\frac{\gamma}{2}t} e^{-i^2\alpha t} e^{i\phi}\\
x(t) = Ae^{-\frac{\gamma}{2}t} e^{-\alpha t} e^{i\phi} + Be^{-\frac{\gamma}{2}t} e^{\alpha t} e^{i\phi}\\
\textrm{Since $A$, $B$, and $\phi$ are constants, we can collapse their products as constants as well. We will call them $A'$ and $B'$}\\
x(t) = A'e^{-\frac{\gamma}{2}t} e^{-\alpha t}+ B'e^{-\frac{\gamma}{2}t} e^{\alpha t}\\
\end{equation}

We can see that here ther isn't really an oscilation. In the first term with $A'$, we have a decaying term with another decaying term, but in the second with $B'$ has a decaying term with a growth term. In the longterm, the larger decaying factor will overpower the mixed decaying and growing term.


## Critical Damping

Critical damping is the case where $w_o^2 - (\frac{\gamma}{2})^2 = 0$, so $\gamma = 2w_o$. We only keep the positive root since by definition, $\gamma$ is a positive value.

This gives us
\begin{equation}
x(t) = Ae^{- \frac{\gamma}{2}t}e^{i\sqrt{w_o^2 - (\frac{\gamma}{2})^2} + \phi}\\
x(t) = Ae^{- \frac{\gamma}{2}t} e^{i \phi}\\
x(t) = A'e^{- \frac{\gamma}{2}t}\\
\end{equation}

So we still don't have an oscillation, but now we don't even have a growth factor. Since this is a second order differential equation, we need another solution. It turns out that $x(t) = Bte^{- \frac{\gamma}{2}t}$ is also a solution. Gougousi gave this to us, just believe it for now.

This means that the complete solution is

$$x(t) = (A' + Bt)e^{-\frac{\gamma}{2}t}$$

So we have 3 different outcomes for our damped harmonic oscillator

### Light Damping

- __Case where $w_o^2 - (\frac{\gamma}{2})^2 > 0$__

This is the only case which gives us an oscillatory motion.
The position is given as 
\begin{equation}
x(t) = Ae^{-\frac{\gamma}{2}t} \cos(wt + \phi)\\
\textrm{where}\\
w = \sqrt{w_o^2 -(\frac{\gamma}{2})^2}\\
\end{equation}

### Heavy Damping

- __Case where $w_o^2 - (\frac{\gamma}{2})^2 < 0$__

There is no oscillatory motion, and we go asymptotically to 0 very quickly. Position is given as 
\begin{equation}
x(t) = Ae^{-\frac{\gamma}{2}t} e^{\alpha t} + Be^{-\frac{\gamma}{2}t} e^{-\alpha t}\\
\textrm{where}\\
\alpha = \sqrt{(\frac{\gamma}{2})^2 - w_o^2}\\
\end{equation}

### Critical Damping

- __Case where $w_o^2 - (\frac{\gamma}{2})^2 = 0$__

This means $\gamma = 2 w_o$ and $x(t) = (A' + Bt)e^{-\frac{\gamma}{2}t}$

No oscillatory motion, and decaying more quickly than heavy damping.

## Energy of a Damped Harmonic Oscillator

The main point whne considering energy of a damped harmonic oscillator is that energy is not conserved, since we are losing amplitude as we move forward. If we take the derivative of energy with respect to time, we will see that it is negative.

\begin{equation}
E = \frac{1}{2}kx^2 + \frac{1}{2}m\frac{dx}{dt}^2\\
\frac{dE}{dt} = kx \frac{dx}{dt} + m\frac{dx}{dt} \frac{d^2x}{dt^2}\\
\frac{dE}{dt} = \frac{dx}{dt} (kx + m\frac{d^2x}{dt^2})\\
\textrm{Remembering that } m\frac{d^2x}{dt^2} + b\frac{dx}{dt} + kx = 0 \textrm{, we can substitute }\\
\frac{dE}{dt} = \frac{dx}{dt} (-b\frac{dx}{dt})\\
\frac{dE}{dt} = -b(\frac{dx}{dt})^2\\
\end{equation}
