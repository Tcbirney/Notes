## The Equation of Motion of a Forced Harmonic Oscillator

### Undamped Forced Oscillations

In the above picutre we cn see the free body diagram of a mass and spring system where there if the applied damping force $F$ which has a value of $F_0 \cos(wt)$. We cna derive the 2nd equation similar to how we have done with the unforced oscillators.

We want to derive an equation for the motion of the forced oscillations with respect to the frequency of the driving force $w$. If we have a force $F_0$ that does not change with time, then we can say that $w$ and acceleration is 0. We can then derive and equation for x when we have low driving frequencies. 

\begin{equation}
m \frac{d^2x}{dt^2} + kx = F_0 \cos(wt)\\
kx = F_0\\
x \rightarrow \frac{F_0}{k}\\
\textrm{as}\\
w \rightarrow 0
\end{equation}


To derive the solution of other forced oscillations let us consider the following scenario where we are holding a rubber band with a mass hanging at the bottom. We then move the rubber band up and down with the equation $\xi = a \cos(wt)$, where a is the amplitude of our motion and w is the frequency. We can the derive that 

\begin{equation}
m \frac{d^2x}{dt^2} = -k(x - \xi)\\
\textrm{And substituting for $\xi$ we get}\\
m \frac{d^2x}{dt^2} +ka= -ka\cos(wt)\\ 
\end{equation}

This has the exact same form as $ka = F_0$ which is when w $\rightarrow$ 0

So at veyr low frequencies the motion is governed by the stiffness of the spring. Then as w is increased and is approaching the resonance frequency, the amplitude if the motion increases. It increases until the resonance frequency is surpasssed and the mass moves in the opposite direction of the driving force. As we increase w the mass once again approaches amplitude 0 since the motion is dominated by the inertia of the block. However, at all frequencies the mass moves up and down periodically at the same frequency as the driving force. This suggests that the displacement can be written as
$$x = A(w)\cos(wt - \delta)$$

Here A(w) is the physical amplitude we observe, and $\delta$ is the difference between the driving force and the displacement. It is negative because we are to assume that displacement lags behind the applied force. We expect $\delta$ to be 0 at very low oscillations since the driving frequency will equal the natural frequency, and $\pi$ at very high frequencies since the mass will move opposite the direction of the applied force.

Substituting $x$ and $\frac{d^2x}{dt^2}$ in for this version of the second law equation gives us

\begin{equation}
m \frac{d^2x}{dt^2} + kx = F_0 \cos(wt)\\
\frac{d^2x}{dt^2} + kA(w)\cos(wt + \delta) = ka \cos(wt + \delta)\\
\frac{d^2x}{dt^2} + w_0^2A(w)\cos(wt + \delta) = w_0^2a \cos(wt + \delta)\\
\frac{d^2x}{dt^2} + w_0^2A(w)(\cos(wt)\cos(\delta) + \sin(wt)\sin(\delta)) = w_0^2a (\cos(wt)\cos(\delta) + \sin(wt)\sin(\delta))\\

\textrm{Then it must be true that}\\

A(w)(1-w^2/w_0^2)\cos(\delta) = a\\
A(w)(1-w^2/w_0^2)\sin(\delta) = 0\\

\end{equation}

When $\delta = 0$, the above equations give
$$A(w) = \frac{a}{(1-w^2/w_0^2)}$$
which proves that w must be < w_0 since A(w) is defined as a positive number.

When $\delta = 0$, the above equations give
$$A(w) = -\frac{a}{(1-w^2/w_0^2)}$$
which proves that w must be > w_0 since A(w) is defined as a negative number.

## Forced Oscillations With Damping

We can use the same equation form as the unforced oscillation except with the net force applied.

\begin{equation}
m \frac{d^2x}{dt^2} + b\frac{dx}{dt}+ kx = F_0 \cos(wt)\\
\frac{d^2x}{dt^2} + \gamma\frac{dx}{dt}+ w_0^2x = \frac{F_0}{m} \cos(wt)\\
\end{equation}

Trying a solution of the form $x = A(w)\cos(wt-\delta)$ and substituing x and its derivatives in the above equation gives us

\begin{equation}
A(w) ((w_0^2 - w^2)\cos(\delta) + w\gamma\sin(\delta)) = w_0^2a\\
\textrm{and}\\
(w_0^2 - w^2)\sin(\delta) = w\gamma\cos(\delta)\\
\textrm{which gives}\\
\tan(\delta) = \frac{w\gamma}{w_0^2 - w^2}
\end{equation}



So as we sweep the values for w form very low oscillations to very high oscillations, gamma moves continuously from 0 to $\pi$, and passes through $\pi/4$ precisely when $w = w_0$. This is consistent with what we have worked out with normal SHO in that $\gamma$ represents the lag betwen the force and displacement and that velocity is always $\pi/4$ ahead in the argand diagram for SHO.

If $\tan(\delta) = \frac{w\gamma}{w_0^2 - w^2}$, then we can say that 
\begin{equation}
\sin(\delta) = \frac{w\gamma}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\

\textrm{and that}\\
\cos(\delta) = \frac{(w_0^2 - w^2)^2}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\

\textrm{which we can then substitute back into the characteristic equation}\\

A(w) ((w_0^2 - w^2)\cos(\delta) + w\gamma\sin(\delta)) = w_0^2a\\\

\textrm{giving}\\
A(w) = \frac{aw_0^2}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\



\end{equation}
