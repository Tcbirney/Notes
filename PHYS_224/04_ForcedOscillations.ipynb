### Undamped Forced Oscillations

In the above picutre we cn see the free body diagram of a mass and spring system where there if the applied damping force $F$ which has a value of $F_0 \cos(wt)$. We cna derive the 2nd equation similar to how we have done with the unforced oscillators.

We want to derive an equation for the motion of the forced oscillations with respect to the frequency of the driving force $w$. If we have a force $F_0$ that does not change with time, then we can say that $w$ and acceleration is 0. We can then derive and equation for x when we have low driving frequencies. We get the 2nd irder non homogeneous differentiable equation

\begin{equation}
m \frac{d^2x}{dt^2} + kx = F_0 \cos(wt)\\
kx = F_0\\
x \rightarrow \frac{F_0}{k}\\
\textrm{as}\\
w \rightarrow 0
\end{equation}

The solution of a non homogenous is the solution to the homogenous equation plus the partial solution to the non homogenours equation. The partial solution is any function that satisfies the Non homogenous. So x = x_1 + x_2 where x_1 = the solution to the non homogenous equation.



To derive the solution of other forced oscillations let us consider the following scenario where we are holding a rubber band with a mass hanging at the bottom. We then move the rubber band up and down with the equation $\xi = a \cos(wt)$, where a is the amplitude of our motion and w is the frequency. We can the derive that 

\begin{equation}
m \frac{d^2x}{dt^2} = -k(x - \xi)\\
\textrm{And substituting for $\xi$ we get}\\
m \frac{d^2x}{dt^2} +ka= -ka\cos(wt)\\ 
\end{equation}

This has the exact same form as $ka = F_0$ which is when w $\rightarrow$ 0

So at veyr low frequencies the motion is governed by the stiffness of the spring. Then as w is increased and is approaching the resonance frequency, the amplitude if the motion increases. It increases until the resonance frequency is surpasssed and the mass moves in the opposite direction of the driving force. As we increase w the mass once again approaches amplitude 0 since the motion is dominated by the inertia of the block. However, at all frequencies the mass moves up and down periodically at the same frequency as the driving force. This suggests that the displacement can be written as
$$x = A(w)\cos(wt - \delta)$$

Here A(w) is the physical amplitude we observe, and $\delta$ is the difference between the driving force and the displacement. It is negative because we are to assume that displacement lags behind the applied force. We expect $\delta$ to be 0 at very low oscillations since the driving frequency will equal the natural frequency, and $\pi$ at very high frequencies since the mass will move opposite the direction of the applied force.

Substituting $x$ and $\frac{d^2x}{dt^2}$ in for this version of the second law equation gives us

\begin{equation}
m \frac{d^2x}{dt^2} + kx = F_0 \cos(wt)\\
\frac{d^2x}{dt^2} + kA(w)\cos(wt + \delta) = ka \cos(wt + \delta)\\
\frac{d^2x}{dt^2} + w_0^2A(w)\cos(wt + \delta) = w_0^2a \cos(wt + \delta)\\
\frac{d^2x}{dt^2} + w_0^2A(w)(\cos(wt)\cos(\delta) + \sin(wt)\sin(\delta)) = w_0^2a (\cos(wt)\cos(\delta) + \sin(wt)\sin(\delta))\\

\textrm{Then it must be true that}\\

A(w)(1-w^2/w_0^2)\cos(\delta) = a\\
A(w)(1-w^2/w_0^2)\sin(\delta) = 0\\

\end{equation}

When $\delta = 0$, the above equations give
$$A(w) = \frac{a}{(1-w^2/w_0^2)}$$
which proves that w must be < w_0 since A(w) is defined as a positive number.

When $\delta = 0$, the above equations give
$$A(w) = -\frac{a}{(1-w^2/w_0^2)}$$
which proves that w must be > w_0 since A(w) is defined as a negative number.

## Forced Oscillations With Damping

We can use the same equation form as the unforced oscillation except with the net force applied.

\begin{equation}
m \frac{d^2x}{dt^2} + b\frac{dx}{dt}+ kx = F_0 \cos(wt)\\
\frac{d^2x}{dt^2} + \gamma\frac{dx}{dt}+ w_0^2x = \frac{F_0}{m} \cos(wt)\\
\end{equation}

Trying a solution of the form $x = A(w)\cos(wt-\delta)$ and substituing x and its derivatives in the above equation gives us

\begin{equation}
A(w) ((w_0^2 - w^2)\cos(\delta) + w\gamma\sin(\delta)) = w_0^2a\\
\textrm{and}\\
(w_0^2 - w^2)\sin(\delta) = w\gamma\cos(\delta)\\
\textrm{which gives}\\
\tan(\delta) = \frac{w\gamma}{w_0^2 - w^2}
\end{equation}



So as we sweep the values for w form very low oscillations to very high oscillations, gamma moves continuously from 0 to $\pi$, and passes through $\pi/4$ precisely when $w = w_0$. This is consistent with what we have worked out with normal SHO in that $\gamma$ represents the lag betwen the force and displacement and that velocity is always $\pi/4$ ahead in the argand diagram for SHO.

If $\tan(\delta) = \frac{w\gamma}{w_0^2 - w^2}$, then we can say that 
\begin{equation}
\sin(\delta) = \frac{w\gamma}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\

\textrm{and that}\\
\cos(\delta) = \frac{(w_0^2 - w^2)^2}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\

\textrm{which we can then substitute back into the characteristic equation}\\

A(w) ((w_0^2 - w^2)\cos(\delta) + w\gamma\sin(\delta)) = w_0^2a\\\

\textrm{giving}\\
A(w) = \frac{aw_0^2}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\

\end{equation}


<div style="text-align: center"><img src="Pictures/a(w)_as_w.png"></div>

Note that in the damped case, A(w) does not $\rightarrow \inf$ as $w \rightarrow 0$, and also, the point of maximum displacement does not occur at $w = w_0$ anymore. This is because the denominiator of the A(w) function must be a minimum. So we must have 

\begin{equation}
\frac{d}{dw}[ (w_0^2-w^2)^2 + w^2\gamma^2]^{1/2}= 0\\
\textrm{Which gives}\\
w  = w_0(1-\gamma^2/2w_0^2)^{1/2} = w_{max}\\
\textrm{And we can then substitue this equation back into the equation for A(w), giving A(max)}\\
A(max) = \frac{aw_0/\gamma}{(1-\gamma^2/4w_0^2)^{1/2}}
\end{equation}


To make a more general solution of the intermediary cases, we can substitue $F_0 = ka$ in the A(w) equation.
\begin{equation}
A(w) = \frac{aw_0^2}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\
A(w) = \frac{F_0/m}{((w_0^2 - w^2)^2 + (w\gamma)^2)^{1/2}}\\
\end{equation}

Damping in systems with forced oscillations can be very importnt to prevent destructive movement. As we have said that without damping, oscillations that encounter a force at resonant frequency have very high amplitudes, so we introduce damping to limit them.

Remember that we have define the quality factor as $Q = \frac{w_0}{\gamma}$, which determines how many oscillations we will have before it dies out. We can substitute Q into the w(max) = equation to get

\begin{equation}
w_{max} = w_0(1-\gamma^2/2w_0^2)^{1/2}\\
w_{max} = w_0(1-1/2Q^2)^{1/2}\\

\textrm{And then into the equation for A(max) to get}\\
A(w) = \frac{aQ}{(1-1/4Q^2)^{1/2}}\\

\end{equation}


For the case of light damping, when Q >> 1, then w_{max} = w_0 and A_{max} = aQ.

## Power Absorbed During Forced Oscillations

When there is no damping, then there is no dissipation of energy. During steady state oscillations, energy is recovered as the spring returns to its equilibrium length. This is why we go to infiinity if we have a resonant force with no damping. 

However, in reality, oscillators lose energy because of th frictional forces present in the system. THe power absorbed by the oscillator is exactly equal to the power dissipated.

As usual for forced damped systems, we will say thet 
\begin{equation}
x = A(w)cos(wt - \delta)\\
\textrm{where}\\
A(w) = \frac{aw_0^2}{\sqrt{(w_o^2-w^2)^2 + w^2\gamma^2}}
\end{equation}

So velocity is then given as 
\begin{equation}
v = \frac{dx}{dt} = -A(w)w\sin(wt-\delta)\\
v = -v_0(w)sin(wt-\delta)\\
\textrm{where}\\
v_0 = A(w)w = \frac{aw_0^2w}{\sqrt{(w_o^2-w^2)^2 + w^2\gamma^2}}\\
\end{equation}

We can consider the $v_0$ as the amplitude of the velocity. We can also see that as $w \rightarrow 0$ and when $w \rightarrow \infty$ then $v_0(w) \rightarrow 0$


Since energy is force over distance, we can calulate the rate of energy loss by multiplying force by the rate of change in distance. So

\begin{equation}
P(t) = bv(t) \times v(t) = b[v(t)]^2\\
= b[v_0(w)]^2\sin^2(wt-\delta)\\
\end{equation}

Since instantaneous power loss varies, we would actually want to look as average power loss over a complete cycle. Which would be

\begin{equation}
\bar{P(w)} = \frac{1}{T} \int_{t_o}^{t_o + T}P(t)dt\\

\bar{P(w)} = \frac{b[v_o(w)]^2}{T} \int_{t_o}^{t_o + T}\sin^2(wt - \delta)dt\\
\textrm{And since this integral when bounded by an oscillitory period is always $\frac{T}{2}$, then}\\

\bar{P(w)} = \frac{b[v_o(w)]^2}{2}\\
\textrm{And substituting in the equation for $v_o$ we get}\\

\bar{P(w)} = \frac{w^2F_o^2\gamma}{2m[(w_o^2-w^2)^2 + w^2\gamma^2]}\\
\end{equation}


This equation for the average power lost by the oscillator gives us the power resonance curve. It shows the power absorbed by the oscillator for any given frequency.

By Approximating $w$ to be close to $w_o$, we can say 
\begin{equation}
(w^2 - w_o^2) = (w_o + w)(w_o - w) = 2w_o(-\Delta w)\\

\textrm{Which gives us the final form of the Power Resonance Curve}\\

\bar{P(w)} = \frac{F_o^2}{2m\gamma (4\Delta w^2/\gamma + 1)}\\
\textrm{And the maximum value of P(w) is given by}
\bar{P_{max}} = \frac{F_o^2}{2m\gamma}
\textrm{And occurs exactly when $w =  w_o$}\\
\end{equation}

The half heights occur when $2\Delta w \gamma = 1$, i.e. when $2\Delta w = \gamma$. Thus the full width at half height is given by, $$w_{fwhh} = 2\Delta w = \gamma = w_o /Q$$
