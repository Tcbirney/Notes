# Coupled Oscillations

## Physcial Charcateristics

The motions of each pendulum are coupled in their oscillatons, so they cannot be onsideredn in isolation.

## Normal Modes

### Mode 1

When we move the two masses in the same direction by the same amplitude.

In this case the force on the spring is not compressed or extended. They oscillate at the frequency of a simple pendulum g/l.

Here x = Acos(wt + phi)

When Starting from rest, the phase angles are zero



### Mode 2

Here we are moving the masses in opposite directions but equal amount,
SO $0 = ma + \frac{mgx}{l} + 2kx$ since the spring i displaced by two times the amount of either mass.

And for small oscillations where $x_a = l\theta_a$, we have 
\begin{equation}
\frac{d^2x}{dt^2} + w^2_2x_a = 0\\
\textrm{where}\\
w_2^2 = \frac{g}{l} + 2\frac{k}{m}\\
\end{equation}

In this case the spring increases restoring force acting on both masses, thus increasing the angular frequency of the oscillation.

So the solution of motion for each mass, a and b, is 
\begin{equation}
x_a = B\cos(w_2t)\\
x_b = B\cos(w_2t + \pi)\\
\end{equation}

### Mode 3: Superposition of Normal Modes

This mode is far more complictaed than the first two. Here the amplitudes of the two masses are not equal, which gives the spring an tension forces of $T = k(x_a - x_b)$

Therefore the restoring force on mass a is $-\frac{mgx_a}{l} -k(x_a -x_b)$ and that of mass b is $-\frac{mgx_a}{l} +k(x_a - x_b)$

and thus the resultant equations of motion are 

\begin{equation}
\frac{d^2x_a}{dt^2} + \frac{gx_a}{l} + \frac{k}{m}(x_a - x_b) = 0\\
\textrm{and}\\
\frac{d^2x_b}{dt^2} + \frac{gx_b}{l} - \frac{k}{m}(x_a - x_b) = 0\\
\end{equation}

Since both of these questions both contain $x_a$ and $x_b$, we cannot solve them seperately, and have to be simultaneously. Adding these equations together gives us
\begin{equation}
\frac{d^2(x_a + x_b)}{dt^2} + \frac{g(x_a + x_b)}{l} = 0\\
\end{equation}


We can see that is equation is of the same form of the equation of simple harmonic motion where the displacement variable is $(x_a + x_b)$ instead of just $x$. Here the frequency of oscillation can be seen to be $\sqrt{\frac{g}{l}}$, which is the same as that of normal mode 1.

If we follow the same procedure, but instead subtract the two solutions instead of adding them, we get 
\begin{equation}
\frac{d^2(x_a - x_b)}{dt^2} + (\frac{g}{l} + \frac{2k}{m})(x_a - x_b) = 0\\
\end{equation}

We can see that the oscillation frequency is $\sqrt{\frac{g}{l} + \frac{2k}{m}}$ which is that same as that of the normal mode 2, and the displacement variable is (x_a - x-b).

We introduce two new variables
\begin{equation}
q_1 = (x_a + x_b)\\
\textrm{and}\\
q_2 = (x_a - x_b)\\
\textrm{then we can say that}\\
\frac{d^2q_1}{dt^2} = w_1^2q1 = 0\\
\textrm{and}\\
\frac{d^2q_2}{dt^2} = w_2^2q2 = 0\\
\end{equation}

This gives us a description of two inependent oscillations in which each normal mode is represented by the oscillation of a single variable.

This gives us the general solutions of these variables of
\begin{equation}
q_1 = C_1\cos(w_1t +\phi_1])\\
\textrm{and}\\
q_2 = C_1\cos(w_2t + \phi_2)\\
\end{equation}

We call $q_1$ and $q_2$ normal corrdinates and $w_1$ and $w_2$ normal frequencies. If $q_1 = 0$, then we care completely in normal mode 2, and thus $x_a = - x_b$ at all times. Likewise if $q_2 = 0$, then we are completely in normal mode 2, and thus $x_a = x_b$ at all times. Since $q_1$ and $q_2$ are completely independent, they do not affect each other in any way and thus energy never flows between normal modes.

We can now express the dispacement of the masses in terms of the independent variables.
\begin{equation}
x_a = \frac{1}{2}(q_1 + q_2)
x_a = \frac{1}{2}[C_1\cos(w_1t + \phi_1) + C_2\cos(w_2t + \phi_2)]

\textrm{and}\\

x_b = \frac{1}{2}(q_1 - q_2)
x_b = \frac{1}{2}[C_1\cos(w_1t + \phi_1) - C_2\cos(w_2t + \phi_2)]
\end{equation}


## Oscillating Masses Coupled By Springs 

### Mode 1
- Where the masses are moving in the same direction

### Mode 2
- Where the masses are moving in opposite directions
- Has higher oscillation frequencies as the additional spring would provide addiional restoring force.

The equations of motion are given as follows 
\begin{equation}
m\frac{d^2x_a}{dt^2} = -kx_a + k(x_b - x_a)\\
m\frac{d^2x_a}{dt^2} = -2ka + kx_b\\
\textrm{and}\\
m\frac{d^2x_b}{dt^2} = -kx_b - k(x_b - x_a)\\
m\frac{d^2x_b}{dt^2} = 2kb + kx_a\\
\end{equation}

We want to find the normal mode of operation where the masses oscillate at the same frequency $w$ so that the general solutions will be $x_a = A\cos(wt)$ and $x_b = B\cos(wt)$

Substituing these general solutions into the equations of motion, then we get 
\begin{equation}
-Amw^2\cos(wt) = kB\cos(wt) - 2kA\cos(wt)\\
\textrm{which gives}\\
\frac{A}{B} = \frac{k}{2k-mw^2}\\

\textrm{And doing the same for mass 2 gives}\\
-Bmw^2\cos(wt) = kA\cos(wt) - 2kB\cos(wt)\\
\textrm{and}\\
\frac{A}{B} = \frac{2k - mw^2}{k}\\
\end{equation}

This tells us that 
\begin{equation}
\frac{A}{B} = \frac{2k - mw^2}{k} = \frac{k}{2k - mw^s}\\
\textrm{And multiplying across gives us}\\
(2k -mw^2)^2 = k^2\\
\end{equation}

This is a quadratic equation where the solution for w is $w^2 = k/m, 3k/m$
