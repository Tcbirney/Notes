## 1.1 Mathematical Modeling of Robots

We will focus on developing the mathematical models that describe robotic systems. These models will allow us to plan and control robot movement.

---

### 1.1.1 Symbolic Representation of Robots

Robots have 
- links
- joints

to form __kinematic chains__

Joints can be 
- rotary (revolute)
    - a hinge that allows for relative rotation about 2 links
    - denoted with __R__
- linear (prismatic)
    - allows linear motion relative to two links
    - denoted with __P__

### 1.1.2 The Configuration Space

Configuration: The complete specification of the loacation of every point on the manipulator

Configuration Space: The set of all possible configurations

We represent a configuration by a set of values for each joint variable.We denote this as a vector $q$ where the joint variables take on the values of $q_1 \dots q_n$, with $q_i = \theta_i$ for a revolute joint and $q_i = d_i$ for a prismatic joint

An object is said to have n degrees of freedom if its configuration can be minimally defined in n paramters. The number of DOFs does not necessarily depend on the number of joints because there may be reduntant joints that do not provide extra degrees of freedom.

A rigid object in 3d space must be dfined by 6 DOFs. X, Y and Z for positioning, and roll, pitch, and yaw for orientation.

As the number of links in a robot increases then the difficulty of controlling the robot increases but so does the reliability.

---

### 1.1.3 The State Space

Configurations provide an instantaneous description of the geometry of the manipulator, but nothing about its dynamic response. 

The __Manipulator State__, on the other hand, when combined with the manipulators dynamics and input, can determine any future state of the manipulator. And the __State Space__ is the set of all possible states. 

For example, in the case of a manipulator arm, the dynamics are newtonian and can be generalized with $F=ma$. And we would be able to represent the state of a joint as $x = (q_i, \dot{q_i})^T$ where $q$ is the value of the joint and $\dot{q}$ is the inear or angular velocity of said joint. Acceleration can be derived programmatically by taking the difference of the velocity over time. The size of the state space is usually $2n$ where the system has $n$ DOFs.

---

### 1.1.4 The Workspace

A robots workspace is the total volume accessible by the manipulator. The workspace is often broken into reachable workspace and dextrous workspace.

Reachable Workspace
- set of points reachable by manipulator

Dextrous Workspace
- set of points reachable given an arbitrary orientation of the end affector

## 1.2 Robots as Mechanical Devices

### 1.2.1 Classification of Robotics Manipulators

Robots can be categrotized by their 
- Power Source
    - Examples are electrical, hydraulic, or pneumatic.
- Joint Actuation Method
- Geometry
    - The geometry of a manipulator is normally characterized by the types of the first 3 joints of the arm. They usually fall into these 5 categories
        - Articulated (RRR)
        - Spherical (RRP)
        - SCARA (RRP)
        - Cylindrical (RPP)
        - Cartesian (PPP)
    - However these are all serially linked models of robots, but we may also have parallel robots.
- Intended Application Area
    - Broken into assembly and non assembly robots. Assembly robots are small electrically driven and either revoulte or SCARA in design.
- Method of Control
    - Broken into servo and non servo robots. Early robots were non servo ad open  loop devices whose movement is limited to predetermined mechanical stops. Servo bots use closed loop computer controls to determine their motion
        - Servo robots are further classified by the method they use to guide the end affector, those being __point to point__ and __continous path__ robots. Point to point robots can be taught a discrete set of points but there is no control on the path the robot takes in between the taught points. Continous path robots have a fully controllable path for the end affector. These are the most advanced.
