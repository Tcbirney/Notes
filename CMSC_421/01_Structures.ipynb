## System Calls

System calls are a __programming interface to the services__ provided by the OS. They are typically written in a high level language such as __C/C++__. They are usually accessed through an Application Programming Interface __(API)__, rather than being called directly. 


The most common APIs are WIN32 for Windows, __POSIX API__ for for POSIX based systems (so all UNIX, LINUX, and MAC OS systems), and JAVA API for for the Java Virtual Machine (JVM)

There are numberous system calls to do something as simple as compying one file to another.

### Implementation

Typycally there is a number associated with each system call, and there is a system call interface that maintains a table indexed according to those numbers. The system call interface is the barrier between user land and kernel land.

### System Call Parameter Passing

A lot of times when calling system calls, more information is required that just the identity of the desired system call. And the information required may vary according to OS and language.

Three general methods to pass parameters to OS

1. Simplest: Pass the parameters in registers. In some cases there may be more paramters than registers, so this wouldn't be feasible

2. Paramters are stored in a block or table in memory, and the address of the block is passed in as a paramter in a register.

3. Paramters are placed, or pushed onto the stack by the program, and popped off the stack by the operating system once they have been read.

The block and stack methods do not limit the number or length of paramters that are passed

### What Kinds of system calls are there?

Process control
- Create, terminate process
- End, abort
- Load, execute
- Get, process attribute
- Wait for time
- Wait event, single event
- Allocate and free memory
- Dump memory if error
- Debugger for determining bugs, single step execution
- Locks for managin access to shared data between processes

File Management
- Create, delete files
- Open, close file
- Read, write, reposition
- get, set file attributes

Device Management
- Request, release device
- Read, write, reposition
- get, set device attributes
- logically attach, detach devices

Information Maintenance
- Get, set time and date
- Get, set system data
- Get, set process, file, or device attributes

Communications
- Create, delete communication connection
- Send, reviece messages
- Shared memory model
- Transfer status information
- Attach, detach remote devices

Protection
- Control access to resources
- Get, set permissions
- Allow, deny user access
