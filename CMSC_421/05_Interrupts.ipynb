# Interrupts

Interrupts are external events that causes a change in the normal execution of the intruction execution and are usually genrerated by hardware devices. Interrupts are asynchronous with respect to the current process. The process can not determine when the next interrupt will happem.

Signals are generated by processes running in the computer.

## Why do we have them

- We want to get the CPUs attention
- around 150 programs each need processing power
- I/O peripherals need attention

But we can't predict when devices need CPU service, so they must be asynchronous. But we want to make sure the CPU stays busy.

## Polling
CPU periodically checks to see if devices need seriving.

Issues
- takes time to poll
- asks fewer times
- inefficient if data is being reported rapidly

USB polls but it has a dedicated chip

## Interrupts

Gives each device a wire that connects directly to the CPU. The device then sends an interrupt signal that gets processed by an interrupt service routine. There is no overhead when there are no requests

Interrupts are better when the processor is not doing other things and response time is not critical

Polling is usually better for real time interrupt handling
- Responds to device immediately
- Fast response time
- Possible to have a second controller

## Hardware Interrupt Handling
Has an interupt controller that signals CPU and passes interrupt number. Cpu Checks for interrupt request after eery executein and if raised then sends it tot he proper ISR. The program stat is saved and then resumed

## SOftware Interrupt Handling
2 parts

__The part that must be done now__
- time sensitive
- must be done in order for device to resume operation

__The part done later__
- To make device faster
- Have more convenient execution

## Interrupt Context

The interrupt service routing uses the context of whatever process was interrupted. That same process state is saved in the original processes structure

The ISR however uses the interrupted threads kernel stack.
- Be careful about stack allocation

The ISR is not allowed to block.
- It has no scheduler
- Cannot call functions that may block

ISRs must be simple and efficient

## Software Interrupts

This is the second part of the ISR. Called bottom halves by Linux.

Things that can be deferred
- Networking
    - Time Critical: Copy packet and respond to hardware
    - Deferred: Process Packet and pass to correct application

- Timers
    - Time Critical: Increment Time
    - Deferred: Recalculate process priorities

We usuaully keep the top half simple and let the longer tasks be executed in the bottom half. The top half is the thing that repsonds to the interrupt and the bottom half is scheduled to be executed later.

### Review

Top Half
- Interrupts Disabled (no interrupt within interrupt unless explicitly enabled)
- Save device data to buffer
- Schedules Bottom Half
- Reenables Interrupts
- Exits

Bottom Half
- Interrupts enabled
- Performs real processing work


## Signals

SOftware Equivalent to hardware interrupts, where we want to interrupt the intruction stream in a process. 

Allows processes to respond to asynchronous external events
- The process may define its own signal handlers
- Defaults
    - Ignore Signal
    - Terminate all threads in process
    - Stop all threads in process
    - Resume all threads in process

These are used as a simple IPC structure

Example: ctrl+c is the SIGINT (2) signal that describes when the user sends an interrupt signal
Which executes

```$ kill -signal pid```

## Signal Basics

Process Structure has flags for possible signals and actions to take. When the signal is posted to the process, the signal pending flag is marked. When the process is next scheduled to run, the pending signals are checked and appropriate action is taken.
- ie signal delivery is not instantaneous

## Review: Signal Vs Interrupts

- __Signals come from processes__
- __Interrupts come form hardware__
- __Both get the attention from the CPU__
- __Something Important Happened that needs work done__
- __Can be split into time critical and deferred sections__
