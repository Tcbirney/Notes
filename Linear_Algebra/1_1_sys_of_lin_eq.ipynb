# Systems of Linear Equations

A __linear equation__ which has variables $x_1$ to $x_n$ can be written in the form 
$$a_1 x_1 + a_2 x_2 + ... + a_n x_n = b$$

where b and the coefficients $a_1 ... a_n$ are real or complex numbers which we usually know in advance.

A __system of linear equations__ is a collection of one or more linear equations involving the same variables. A __solution__ of the system is a list ($s_1, s_2, ... s_n$) of numbers that makes each equation a true statement where the solution is substituted for each $x$. A __solution set__ is the set of all possible solutions to a system of linear equations. And two linear equations ae considered __equivalent__ if they have the same solution set.

Finding the solution set amounts to finding the intersection of the lines represented by the equations in the system.

A system of equations has

- No solutions if the lines represented by the equations do not ever intersect

- One or more solutions if the lines have a finite amount of intersection points

- Infinitely many solutions if they have infinitely many intersection points. This means that the equations in the system are __linearly equivalent__.

A system of Equations is __consistent__ if it has at least one solution, and it is considered inconsistent if it has no solutions.


## Matrix Notation

Lets say we have these linear equations

$$ x_1 - 2x_2 + x_3 =  0$$
$$ 2x_2 - 8x_3 =  8$$
$$ 5x_1 - 5x_3 = 10$$

This can be put into augmented matrix form as

\begin{bmatrix}
1 & -2 & 1 & 0\\
0 & 2 & -8 & 8\\
5 & 0 & -5 & 10
\end{bmatrix}

Where the first 3 columns in the matrix represent each variable, and the final row represents the right hand side of the equation.

The size of the matrix tells us how many rows or columns it has. An $m x n$ matrix is a rectangular array with $m$ rows and $n$ columns

## Solving a Linear System

The basic strategy is to replace one system with an equivalent system, that is easier to solve.

The three basic operations to simplify a linear system are

- Replace one equation by the sum of itself and a multiple of another equation

- Interchange two equations

- Multiply all elements of one equation by a nonzero constant

Lets have the system 

\begin{bmatrix}
1 & -2 & 1 & 0\\
0 & 2 & -8 & 8\\
5 & 0 & -5 & 10
\end{bmatrix} 


We can simplify this by subtracting row 3 by 5 * row 1.

This gives us

\begin{bmatrix}
1 & -2 &  1 & 0\\
0 &  2 & -8 & 8\\
0 &  -10 & -10 & 10
\end{bmatrix}

Which we can see removes the x1 coefficient from row 3. We can use a similar process to remove the x2 coefficient from row 3. We can add 5 * row 2 to row 3.


\begin{bmatrix}
1 & -2 &  1 & 0 \\
0 &  2 & -8 & 8 \\
0 &  0 & -50 & 50 
\end{bmatrix}


And now we can divide rows 2 and 3 to get the leading coefficient to be 1. This matrix is now in triangular form, which is good because it allows us to solve each equation indepentendtly.

\begin{bmatrix}
1 & -2 &  1 & 0 \\
0 &  1 & -4 & 4 \\
0 &  0 & 1 & -1 
\end{bmatrix}

We work our way from the bottom. We can see the bottom rows equation as $1x_3 = -1$ and the only solution is $x_3 = -1$. Now we plug that into the equation for row 2 whcih is $1 x_2 - 4 x_3 = 4$ and when plugging in $x_3$ we get $x_2 + 4 = 0$, then $x_2$ must be $0$. Finally we can plug those variables into the equation of row 1, giving us $x_1 + 1(-1) = 0$, so $x_1 = 1$ 

Therefore the solution to this system of equations is $(1, 0, -1)$



## Existence and Uniqueness

### The Two Fundamental Questions About A Linear System

1. Is the sustem __consistent__; that is does at least one solution _exist_?

2. If a solution exists, is it the _only_ one, that is, is the solution __unique__?

### Determine if the following matrix is consistent

\begin{bmatrix}
1 & -2 & 1 & 0\\
0 & 2 & -8 & 8\\
5 & 0 & -5 & 10
\end{bmatrix}

From before we know that this matrix simplifies to

\begin{bmatrix}
1 & -2 &  1 & 0 \\
0 &  1 & -4 & 4 \\
0 &  0 & 1 & -1 
\end{bmatrix}

Which has the solution $(1, 0, -1)$, so this system is consistent, and it is unique since there are no free variables.

A matrix will fail to be consistent if there is an equation which is contradictory and impossible to satisfy.
